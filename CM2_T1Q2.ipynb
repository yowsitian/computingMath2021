```
import matplotlib.pyplot as plt
import numpy as np 
```

```
# naming the x axis 
plt.xlabel('Width') 
# naming the y axis 
plt.ylabel('Material Cost') 
# giving a title to graph 
plt.title('Material Cost Function, C(w)') 

# setting the x - coordinates 
w = np.arange(0, 50, 1) 
# setting the corresponding y - coordinates 
c = (20*w**3+30)/w
  
# potting the points 
plt.plot(w, c) 
  
# function to show the plot 
plt.show() 
```

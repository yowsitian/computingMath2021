```
import matplotlib.pyplot as plt
import numpy as np 
```

```
# naming the x axis 
plt.xlabel('x') 
# naming the y axis 
plt.ylabel('y') 
# giving a title to graph 
plt.title('y=|x^2-1|') 

# setting the x - coordinates 
x = np.arange(0, 10, 2) 
# setting the corresponding y - coordinates 
y = np.abs(x**2-1)
  
# potting the points 
plt.plot(x, y) 
plt.grid('True')
  
# function to show the plot 
plt.show() 
```
